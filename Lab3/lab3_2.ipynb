#### IDW / Kriging / Global Polynomial Interpolation

```python
def interpolation_methods(feature_class, values):
    for value in values:
        IDW = Idw(feature_class, value)
        IDW.save(f"IDW_{value}")
        
        Kriging_tool = Kriging(feature_class, value, KrigingModelOrdinary())
        Kriging_tool.save(f"Kriging_{value}")
        
        GPI = arcpy.ga.GlobalPolynomialInterpolation(feature_class, value, out_raster=f"GPI_{value}")


interpolation_methods("TMP_stations", ["MEAN_Avg", "MAX_MAX", "MIN_MIN"])
```
